## Diferencia entre vulnerabilidad, amenaza y riesgo

## Vulnerabilidad: 
- Es una debilidad o fallo en un sistema, proceso o control que podría ser explotado, permitiendo que ocurra un incidente de seguridad.

Ejemplo:
- Un control de acceso físico deficiente en el datacenter permite que personas no autorizadas ingresen sin supervisión.
- Software sin actualizar que contiene fallas conocidas, lo cual podría permitir que un atacante tome control del sistema.

## Amenaza: 
- Es cualquier acción, evento o circunstancia que pueda aprovechar una vulnerabilidad para causar daño, pérdida de información, interrupción de servicios o un incidente de seguridad.

### Tipos de Amenazas:
1. Amenazas naturales: Provienen de fenómenos naturales, fuera del control humano.
- Ejemplo: Inundaciones, terremotos, incendios forestales.

2. Amenazas antrópicos: Son causadas por actividades humanas.
- Ejemplo: Protestas, vandalismo, accidentes laborales o de transporte.

3. Amenazas tecnológicas: Se originan por fallos en la tecnología o por ataques dirigidos al sistema.
- Ejemplo: Ataques de Denegación de Servicios (DoS), en los cuales un atacante satura el sistema con tráfico hasta interrumpir el servicio.

Tipos de amenazas tecnológicas:
- Intencionales: Amenazas específicamente diseñadas para comprometer la seguridad de los sistemas.
Ejemplo: Malware (virus, troyanos, ransomware) diseñado para robar datos o causar daño.

- No intencionales: Son incidentes de seguridad que ocurren sin intención maliciosa.
Ejemplo: Un empleado olvida bloquear su equipo antes de salir o no actualiza el sistema, exponiendo el software a fallas.

## Riesgo: 
Es la probabilidad de que una amenaza explote una vulnerabilidad, causando un impacto negativo en la organización, ya sea financiero, reputacional, operativo o en la integridad de la información.

Ejemplo: Interbank presencio un robo de datos de clientes debido a una combinación de amenazas y vulnerabilidades. Si los datos de clientes no estaban cifrados adecuadamente (vulnerabilidad), un atacante podría aprovechar esta debilidad para acceder a información confidencial mediante acceso que le da el intruso dentro de la organización (amenaza), resultando en el robo de datos personales (riesgo materializado).
