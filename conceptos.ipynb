## Diferencia entre vulnerabilidad, amenaza y riesgo

## Vulnerabilidad: 
- Es una debilidad o fallo en un sistema, proceso o control que podra ser explotado, permitiendo que ocurra un incidente de seguridad.

Ejemplo:
- Un control de acceso fsico deficiente en el datacenter permite que personas no autorizadas ingresen sin supervisin.
- Software sin actualizar que contiene fallas conocidas, lo cual podra permitir que un atacante tome control del sistema.

## Amenaza: 
- Es cualquier accin, evento o circunstancia que pueda aprovechar una vulnerabilidad para causar dao, prdida de informacin, interrupcin de servicios o un incidente de seguridad.

### Tipos de Amenazas:
1. Amenazas naturales: Provienen de fenmenos naturales, fuera del control humano.
- Ejemplo: Inundaciones, terremotos, incendios forestales.

2. Amenazas antrpicos: Son causadas por actividades humanas.
- Ejemplo: Protestas, vandalismo, accidentes laborales o de transporte.

3. Amenazas tecnolgicas: Se originan por fallos en la tecnologa o por ataques dirigidos al sistema.
- Ejemplo: Ataques de Denegacin de Servicios (DoS), en los cuales un atacante satura el sistema con trfico hasta interrumpir el servicio.

Tipos de amenazas tecnolgicas:
- Intencionales: Amenazas especficamente diseadas para comprometer la seguridad de los sistemas.
Ejemplo: Malware (virus, troyanos, ransomware) diseado para robar datos o causar dao.

- No intencionales: Son incidentes de seguridad que ocurren sin intencin maliciosa.
Ejemplo: Un empleado olvida bloquear su equipo antes de salir o no actualiza el sistema, exponiendo el software a fallas.

## Riesgo: 
Es la probabilidad de que una amenaza explote una vulnerabilidad, causando un impacto negativo en la organizacin, ya sea financiero, reputacional, operativo o en la integridad de la informacin.

Ejemplo: Interbank presencio un robo de datos de clientes debido a una combinacin de amenazas y vulnerabilidades. Si los datos de clientes no estaban cifrados adecuadamente (vulnerabilidad), un atacante podra aprovechar esta debilidad para acceder a informacin confidencial mediante acceso que le da el intruso dentro de la organizacin (amenaza), resultando en el robo de datos personales (riesgo materializado).
